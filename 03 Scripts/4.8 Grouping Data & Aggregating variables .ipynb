### 4.Aggregating Data with transform()

### Deriving Columns with loc()

```python
### create a flag that assigns a loyalty 
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
### print the frequency of the new â€œloyalty_flagâ€ column using the value_counts() function:
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# 5. whether thereâ€™s a difference between the spending habits of the three types of customers

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
### To find the difference between the spending habits of the three types of customers i find the mean, min, max value for the 
### loyal customers . The mean value for the loyal customer is low when compare to other two customers. 
### This shows the prices of products purchased by loyal customers differ from those purchased by regular or new customers.
```

```python
df_ords_prods_merge['prices_mean'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

# 6. Different types of spenders in their marketing campaigns.

```python
### create a spending flag 
df_ords_prods_merge.loc[df_ords_prods_merge['prices_mean'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices_mean'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 7.Determine frequent versus non-frequent customers. 

```python
df_ords_prods_merge['days_since_prior_order_median'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['days_since_prior_order_median'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['days_since_prior_order_median'] > 10) & (df_ords_prods_merge['days_since_prior_order_median'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['days_since_prior_order_median'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
