## 4.Create a bar chart from the loyalty_flag column

```python
# Group by region and calculate order count and sum of prices
order_stats = df_ords_prods_custm.groupby('loyalty_flag').agg({
    'order_id': 'count',
    'prices': 'sum'
}).reset_index()

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.35
bar_positions = range(len(order_stats))

# Bar chart for order count
order_count_bar = ax.bar(bar_positions, order_stats['order_id'], width=bar_width, label='Order Count', color='skyblue')

# Bar chart for sum of prices
sum_of_prices_bar = ax.bar([pos + bar_width for pos in bar_positions], order_stats['prices'], width=bar_width, label='Sum of Prices', color='orange')

# Annotate the bars with actual values
for bars, values in zip([order_count_bar, sum_of_prices_bar], [order_stats['order_id'], order_stats['prices']]):
    for bar, value in zip(bars, values):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{value}",
                ha='center', va='bottom', color='black')

ax.set_xlabel('Loyalty')
ax.set_ylabel('Count / Sum')
ax.set_xticks([pos + bar_width / 2 for pos in bar_positions])
ax.set_xticklabels(order_stats['loyalty_flag'])
ax.legend()

plt.title('Loayalty Order Count and Sum of Prices')
plt.savefig(os.path.join(path, '04 Analysis','Visualization', 'bar_loyalt_flag.png'))
plt.show()
```
